#  GISAXS (Grazing-incidence small angle X-ray scattering geometry)

### More details skbeam.core.recip module: gisaxs function.
### (https://github.com/scikit-beam/scikit-beam/blob/master/skbeam/core/recip.py)

```python
import skbeam.core.recip as recip

import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

## get the inputs

```python
inc_x0 = 1871   # incident beam x component in pixels
inc_y0 = 339    # incident beam y component in pixels
refl_x0 = 1871   # refelcted beam x component in pixels
refl_y0 = 811    # refelcted beam y component in pixels    
dist_sample = 4.81  # distance to the sample in mm
wavelength = 1.3776 #  wavelength of the X-ray beam in Angstroms

dpix = 7.5000004e-02  # pixel size
```

```python
pixel_size = (dpix, dpix)
incident_beam = (inc_x0, inc_y0)
reflected_beam = (refl_x0, refl_y0)
detector_size = (2167, 2070)
```

```python
g_output = recip.gisaxs(incident_beam, reflected_beam, pixel_size,
                        detector_size, dist_sample, wavelength,
                        theta_i=0.0)
```

### Incident Angle

```python
g_output.alpha_i    
```

## Tilt Angle

```python
g_output.tilt_angle
```

## Exit Angle

```python
fig, ax = plt.subplots(figsize=(6, 6))
im = ax.imshow(g_output.alpha_f, cmap="viridis")
ax.set_title("Exit Angle")
fig.colorbar(im)
plt.show()
```

## X component of the scattering wave vector

```python
fig, ax = plt.subplots(figsize=(6, 6))
im = ax.imshow(g_output.qx, cmap="viridis")
ax.set_title("X component of the scattering wave vector")
fig.colorbar(im)
plt.show()
```

## Y component of the scattering wave vector

```python
fig, ax = plt.subplots(figsize=(6, 6))
im = ax.imshow(g_output.qy, cmap="viridis")
ax.set_title("Y component of the scattering wave vector")
fig.colorbar(im)
plt.show()
```

## Z component of the scattering wave vector

```python
fig, ax = plt.subplots(figsize=(6, 6))
im = ax.imshow(g_output.qz, cmap="viridis")
ax.set_title("Z component of the scattering wave vector")
fig.colorbar(im)
plt.show()
```

## Scattering Wave Vector Parallel Component

```python
fig, ax = plt.subplots(figsize=(6, 6))
im = ax.imshow(g_output.qr, cmap="viridis")
ax.set_title("Scattering Wave Vector : Parallel Component")
fig.colorbar(im)
plt.show()
```
